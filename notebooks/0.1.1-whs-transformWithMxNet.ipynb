```python
import mxnet as mx
from mxnet.gluon.model_zoo import vision
import os
import numpy as np
import glob
import pandas as pd
#from scipy.spatial.distance import cosine
from IPython.display import Image 

# set the context on CPU, switch to GPU if there is one available
ctx = mx.cpu()

import glob2
from tqdm import tqdm
```

```python
%time densenet = vision.densenet201(pretrained=True)
type(densenet)
```

```python
#all_header_files = glob2.glob('src/**/*.jpg')
fn_df_save = os.path.join(os.path.dirname(os.getcwd()), 'data', 'processed','0.0.4-whs-dogVectors.pickle')
fn_df_load = os.path.join(os.path.dirname(os.getcwd()), 'data', 'processed','0.0.3-whs-dogVectors.pickle')
if not os.path.isdir(os.path.dirname(fn_df_save)):
    os.mkdir(os.path.dirname(fn_df_save))
try:
    df_load = pd.read_pickle(fn_df_save)
except:
    pass

df = pd.DataFrame()
df['fn'] = None
df['vector'] = None
fns = glob2.iglob(os.path.join(os.path.dirname(os.getcwd()), 'data', 'raw','**/*.jpg'))

for index, fn in enumerate(tqdm(fns)):
    try:
        #print('Processing: ',fn)
        if os.path.isdir(fn):
            fn = glob.iglob(os.path.join(fns, '*'))
        df.loc[index,'fn'] = fn
        image = mx.image.imdecode(open(fn, 'rb').read()).astype(np.float32)
        resized = mx.image.resize_short(image, 224) #minimum 224x224 images
        cropped, crop_info = mx.image.center_crop(resized, (224, 224))
        normalized = mx.image.color_normalize(cropped/255,
                                              mean=mx.nd.array([0.485, 0.456, 0.406]),
                                              std=mx.nd.array([0.229, 0.224, 0.225])) 
        # the network expect batches of the form (N,3,224,224)
        flipped_axis = normalized.transpose((2,0,1))  # Flipping from (224, 224, 3) to (3, 224, 224)
        batchified = flipped_axis.expand_dims(axis=0) # change the shape from (3, 224, 224) to (1, 3, 224, 224)
        df.loc[index,'vector'] = densenet.features(batchified)[0].asnumpy()
        if index % 100 == 0:
            df_full = pd.concat([df_load,df])
            df_full.to_pickle(fn_df_save)
    except:
        pass

df.to_pickle(fn_df_save)
df.tail()
```
